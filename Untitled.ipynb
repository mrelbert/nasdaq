```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
stock_list = pd.read_csv('./nasdaq_12042020.csv')
```

```python
tickers = stock_list.dropna(subset=['Last Sale']) 
```

```python
tickers['Last Sale'] = tickers['Last Sale'].str.replace("$", "")
tickers['Last Sale'] = tickers['Last Sale'].astype(float)
```

```python
tickers.shape
```

```python
tickers.sort_values(by=['Last Sale'], ascending=False)
tickers.head()
```

```python
five = tickers.loc[tickers['Last Sale'] < 5]
```
